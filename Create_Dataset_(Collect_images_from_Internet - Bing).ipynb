```python
from googlesearch import search
import urllib.request

# query to search for
query = "train bogie images"

# maximum number of images to download
num_images = 10

# loop through the Google search results and download images
for url in search(query, stop=num_images):
    try:
        # open the URL and read the contents
        with urllib.request.urlopen(url) as response:
            content = response.read()

        # save the contents to a file
        with open(f"image{num_images}.jpg", "wb") as f:
            f.write(content)
        
        # increment the image counter
        num_images += 1

    except:
        print(f"Unable to download image {num_images} from {url}")
```
